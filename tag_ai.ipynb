```python


prompts = []
completions = []

for _, row in df_top.iterrows():
    tags_list = ast.literal_eval(row['tags'])

    prompt_text = f"Generate a YouTube title and description for a computer science/Data Science Shorts video based on tags: {tags_list}."
    completion_text = f"Title: {row['video_title']} \nDescription: {row['video_description']}"

    prompts.append(prompt_text)
    completions.append(completion_text)


df_prompt_completion = pd.DataFrame({
    'prompt': prompts,
    'completion': completions
})

df_prompt_completion.head()
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_name = "HuggingFaceTB/SmolLM2-135M-Instruct"
device = "cuda"  # for GPU usage or "cpu" for CPU usage

# load the tokenizer and the model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    dtype=torch.float32,
    low_cpu_mem_usage=True,
).to(device)
```

```python
from transformers import pipeline
prompt = "give me  a trending youtube video title about computer science"
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, device=0 if device == "cuda" else -1)
print(pipe(prompt, max_new_tokens=200))
```

```python
from peft  import LoraConfig, get_peft_model
lora_config = LoraConfig(
    r=4,
    lora_alpha=32,
    target_modules=["q_proj", "v_proj"],
    lora_dropout=0.15,
    bias="none",
    task_type="CAUSAL_LM"
)
peft_model = get_peft_model(model, lora_config)



```

```python
from datasets import load_dataset

# Load your JSONL dataset
dataset = load_dataset('json', data_files='new_youtube_cs_shorts.json')

def tokenize_prompt_completion(example, max_length=512):
    # Tokenize prompt and completion separately (get only input_ids)
    prompt = tokenizer(example['prompt'], max_length=max_length, add_special_tokens=False)
    completion = tokenizer(example['completion'],max_length=max_length, add_special_tokens=False)

    prompt_ids = prompt['input_ids']
    completion_ids = completion['input_ids']

    # Combine prompt + completion
    input_ids = prompt_ids + completion_ids
    attention_mask = [1] * len(input_ids)

    # Labels: mask prompt tokens, learn on completion
    labels = [-100] * len(prompt_ids) + completion_ids

    # Truncate if necessary
    input_ids = input_ids[:max_length]
    attention_mask = attention_mask[:max_length]
    labels = labels[:max_length]

    # Pad to max_length
    pad_length = max_length - len(input_ids)
    if pad_length > 0:
        input_ids += [tokenizer.pad_token_id] * pad_length
        attention_mask += [0] * pad_length
        labels += [-100] * pad_length

    return {
        "input_ids": input_ids,
        "attention_mask": attention_mask,
        "labels": labels
    }

# Apply to dataset
tokenized_dataset = dataset['train'].map(
    tokenize_prompt_completion,
    remove_columns=dataset['train'].column_names
)

print(tokenized_dataset[0])
```

```python
from transformers import Trainer, TrainingArguments
training_args = TrainingArguments(
    output_dir="./smollm2-135m-instruct-cs-shorts-finetuned",
    per_device_train_batch_size=2,
    gradient_accumulation_steps=4,
    num_train_epochs=3,
    learning_rate=2e-4,
    logging_steps=10,
    save_steps=50,
    save_total_limit=2,
    fp16=True,
    save_strategy="epoch",
    lr_scheduler_type="cosine",
    report_to=[]
)
trainer = Trainer(
    model=peft_model,
    args=training_args,
    processing_class=tokenizer,
    train_dataset=tokenized_dataset

)

trainer.train()
```

```python
def generate_text(prompt, max_tokens=300):
    inputs = tokenizer(prompt, return_tensors="pt").to(device)
    output_sequences = model.generate(
        **inputs,
        max_new_tokens=max_tokens,
        temperature=0.6,
        top_p=0.95,
        do_sample=True
    )
    generated_text = tokenizer.decode(output_sequences[0], skip_special_tokens=True)
    return generated_text
```

```python
def generate_text(prompt, max_tokens=200, temperature=0.6, top_p=0.95):
    gen_pipe = pipeline("text-generation", model=peft_model, tokenizer=tokenizer,temperature=0.5)
    output = gen_pipe(prompt, max_new_tokens=max_tokens, temperature=temperature, top_p=top_p)
    return output[0]['generated_text']
```
