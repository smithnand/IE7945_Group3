```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cleaned_youtube_trending_data.csv', parse_dates=['trending_date','video_published_at', 'channel_published_at'])
df.head()
```

```python
df.columns
```

```python
rank_bins = [0,10,20,30,40,50]
rank_lables = ['top10', 'top11-20', 'top21-30', 'top31-40', 'top41-50']
df['rank_range'] = pd.cut(df['rank'], bins=rank_bins, labels=rank_lables, right=True)
df.info()
```

```python
df['rank_range'].value_counts()
```

```python
hist_cols = ['video_view_count',
 'video_like_count',
 'video_comment_count',
 'channel_view_count',
 'subscriber_count',
 'video_count',
 'duration_min'
 ]

for col in hist_cols:
    fig, axes = plt.subplots(1,2,figsize=(12,5))
    sns.histplot(data=df, x=np.log10(df[col]+1), bins=20, kde=True, palette='Set2', ax=axes[0])
    axes[0].set_title(f'{col} distribution (Log10)')

    sns.histplot(data=df, x=np.log10(df[col]+1), bins=20, kde=True, palette='Set2', hue=df['rank_range'], ax=axes[1])
    axes[1].set_title(f'{col} distribution (Log10) by rank range')
    plt.suptitle(f'Log10 distribution of {col}', fontsize=15)
    plt.show()
```

3-5min most popular

```python
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

no linear relationship between rank and video performance, channel performance or duration

```python

sns.countplot(data=df, y='category_title', palette='Set2',order=df['category_title'].value_counts().index)
plt.title(f'Count plot of category', fontsize=15)
plt.show()
```

```python
metric_cols = ['video_view_count', 'video_like_count', 'video_comment_count']
top_video_categories = {}

for col in metric_cols:
    top_video_categories[col] = df.groupby('category_title')[col].mean().sort_values(ascending=False)
    sns.barplot(y=top_video_categories[col].index, x=top_video_categories[col].values, palette='Set2')
    plt.title(f'Top 10 Categories by Average {col}', fontsize=15)
    plt.xlabel(f'Average{col}')
    plt.ylabel('Category')
    plt.show()
```

```python
df['day of week']= df['video_published_at'].dt.day_name()
publish_trend = df['day of week'].value_counts()
publish_trend
```

```python
sns.barplot(y=publish_trend.index, x=publish_trend.values, palette='Set2')
plt.title(f'Publish trend by day of the week', fontsize=15)
plt.xlabel('Number of videos published')
plt.ylabel('Day of the week')
plt.show()
```

```python
df['hour of day']= df['video_published_at'].dt.hour.astype(str)
publish_trend_hour = df['hour of day'].value_counts().sort_values(ascending=False).head(10)
publish_trend_hour
```

```python
sns.barplot(y=publish_trend_hour.index, x=publish_trend_hour.values, palette='Set2')
plt.title(f'Publish trend by hour of the day (top10)', fontsize=15)
plt.xlabel('Number of videos published')
plt.ylabel('hour of the day')
plt.show()
```

```python
channel_total_views = df.groupby('channel_title')['channel_view_count'].max().sort_values(ascending=False).head(10)
channel_total_views
```

```python
sns.barplot(y=channel_total_views.index, x=np.log10(channel_total_views.values), palette='Set2')
plt.title(f'Top10 channels by channel total views (Log10)', fontsize=15)
plt.xlabel('Number of channel total views (Log10)')
plt.ylabel('Channel title')
plt.show()
```

```python
channel_subscribers = df.groupby('channel_title')['subscriber_count'].max().sort_values(ascending=False).head(10)
channel_subscribers
```

```python
sns.barplot(y=channel_subscribers.index, x=np.log10(channel_subscribers.values), palette='Set2')
plt.title(f'Top10 channels by subscribers (Log10)', fontsize=15)
plt.xlabel('Number of channel subscribers')
plt.ylabel('Channel title')
plt.show()
```

```python
channel_average_views_per_video = (df.groupby('channel_title')['channel_view_count'].max()/df.groupby('channel_title')['video_count'].max()).sort_values(ascending=False).head(10)
channel_average_views_per_video
```

```python
sns.barplot(y=channel_average_views_per_video.index, x=np.log10(channel_average_views_per_video.values), palette='Set2')
plt.title(f'Top10 channels by average views per video (Log10)', fontsize=15)
plt.xlabel('Average of views per video (Log10)')
plt.ylabel('Channel title')
plt.show()
```
