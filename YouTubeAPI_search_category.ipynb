```python
# import liabraries
import requests
import pandas as pd
import numpy as np
import time
from datetime import datetime
```

```python
API_KEY = 'xxxxxxx'
BASE_URL = "https://www.googleapis.com/youtube/v3"
```

```python
# ========================
# STEP 2: DEFINE SEARCH PARAMETERS
# ========================

# parameters for United States
params_us = {
    'part': 'snippet',
    # 'maxResults': 50,
    'regionCode': 'US', # Change region code as needed
    'key': API_KEY
    }

# parameters for Canada
params_ca = {
    'part': 'snippet',
    # 'maxResults': 50,
    'regionCode': 'CA', # Change region code as needed
    'key': API_KEY
    }
```

```python
# ========================
# STEP 3: CALL YOUTUBE API
# ========================
response = requests.get(f"{BASE_URL}/videoCategories", params=params_us)
if response.status_code == 200:
    data = response.json()
else:
    print("Error:", response.status_code, response.text)
    data = None
```

```python
# ========================
# STEP 4: PARSE DATA INTO DATAFRAME
# ========================
if data:
    category_dict_us = {}
    for i, item in enumerate(data.get('items', [])):
        category = {
            
            'category_id': item['id'],
            'category_title': item['snippet']['title'],
            }
        category_dict_us[category['category_id']] = category['category_title']

    
    print(category_dict_us)
else:
    print("No data fetched.")
```
